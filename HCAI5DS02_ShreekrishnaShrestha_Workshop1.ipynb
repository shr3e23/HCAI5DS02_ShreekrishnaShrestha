```python
import numpy as np
import pandas as pd

sprint_times = np.array([5.1, 4.8, 5.0, 4.9, 5.3, 4.7, 4.9, 4.8, 5.2, 5.0])

mean_time = np.mean(sprint_times)
std_dev_time = np.std(sprint_times, ddof=1)

print(f"Mean sprint time: {mean_time:.2f} seconds")
print(f"Sample Standard Deviation: {std_dev_time:.2f} seconds")

coefficient_of_variation = (std_dev_time / mean_time) * 100

print(f"Coefficient of Variation: {coefficient_of_variation:.2f}%")

print("\nAre the player's times tightly clustered or highly variable?")
if coefficient_of_variation < 30:
    print("The player's times are relatively tightly clustered.")
else:
    print("The player's times are relatively highly variable.")

print("\nImpact of an outlier:")
outlier_sprint_times = np.array([5.1, 4.8, 5.0, 4.9, 5.3, 4.7, 4.9, 4.8, 5.2, 3.8])  #Replacing 5.0 with 3.8
mean_time_outlier = np.mean(outlier_sprint_times)
std_dev_time_outlier = np.std(outlier_sprint_times, ddof=1)
coefficient_of_variation_outlier = (std_dev_time_outlier / mean_time_outlier) * 100

print(f"Mean sprint time with outlier: {mean_time_outlier:.2f} seconds")
print(f"Sample Standard Deviation with outlier: {std_dev_time_outlier:.2f} seconds")
print(f"Coefficient of Variation with outlier: {coefficient_of_variation_outlier:.2f}%")
```


The mean is the average of all data points, while the median is the middle value when the data is ordered. The mean is heavily influenced by extreme values (outliers), whereas the median is not.

If Tracker B's step count data has a higher mean but a lower median, this indicates that the distribution is likely right-skewed. In a right-skewed distribution, there is a tail of higher values (potentially outliers) that pulls the mean towards the higher end, while the median remains closer to the bulk of the data.

This skewness could be due to various factors in fitness tracker data, such as occasional bursts of very high step counts (e.g., during intense activity or due to errors in measurement) that are significantly higher than the typical daily step counts.

In summary, the presence of higher-than-average step counts in Tracker B's data is likely inflating the mean, causing it to be higher than the median, which is less affected by these extreme values.

Interquartile Range (IQR) and Standard Deviation (SD) provide complementary insights into data variability.

IQR measures the spread of the middle 50% of the data. It is the difference between the third quartile (Q3, 75th percentile) and the first quartile (Q1, 25th percentile). The IQR is a robust measure of spread, meaning it is not heavily influenced by extreme values or outliers.

Standard Deviation measures the typical deviation or distance of each data point from the mean. It provides a measure of the overall spread of the data, taking into account all data points. Unlike the IQR, the standard deviation is sensitive to extreme values and outliers.

Comparing the IQR and SD can reveal information about the shape of the distribution and the presence of outliers:
- In a perfectly symmetrical distribution (like a normal distribution), the SD is approximately 1.345 times the IQR (SD  1.345 * IQR).
- In a skewed distribution or one with outliers, the relationship between IQR and SD will differ. The SD will be disproportionately larger compared to the IQR because outliers have a greater impact on the calculation of the SD than the IQR.

Considering Tracker B, we previously noted that its mean was higher than its median, suggesting a right-skewed distribution or the presence of outliers. If Tracker B also has a standard deviation that is relatively large compared to its IQR (e.g., if SD > 1.345 * IQR), this would further support the conclusion of skewness or the presence of outliers in its step count data.

In summary, while IQR shows the spread of the central part of the data, SD reflects the overall spread, including the influence of extreme values. A significant difference in their relationship compared to a symmetrical distribution, particularly a much larger SD relative to the IQR, can indicate skewness or outliers, aligning with the observation of a mean higher than the median.

```python
import numpy as np
antibody_levels_under_40 = np.array([840, 800, 140, 500, 1100])
antibody_levels_over_60 = np.array([620, 610, 85, 480, 800])

# Calculate standard deviation
std_dev_under_40 = np.std(antibody_levels_under_40, ddof=1)
std_dev_over_60 = np.std(antibody_levels_over_60, ddof=1)

# Calculate mean
mean_under_40 = np.mean(antibody_levels_under_40)
mean_over_60 = np.mean(antibody_levels_over_60)

# Calculate coefficient of variation
cv_under_40 = (std_dev_under_40 / mean_under_40) * 100
cv_over_60 = (std_dev_over_60 / mean_over_60) * 100

print(f"Under 40 Group:")
print(f"  Mean Antibody Level: {mean_under_40:.2f}")
print(f"  Standard Deviation: {std_dev_under_40:.2f}")
print(f"  Coefficient of Variation: {cv_under_40:.2f}%")
print("\nOver 60 Group:")
print(f"  Mean Antibody Level: {mean_over_60:.2f}")
print(f"  Standard Deviation: {std_dev_over_60:.2f}")
print(f"  Coefficient of Variation: {cv_over_60:.2f}%")

print("\nComparison of Spread:")
if cv_under_40 > cv_over_60:
    print("The Under 40 group shows relatively more spread in antibody levels.")
elif cv_over_60 > cv_under_40:
    print("The Over 60 group shows relatively more spread in antibody levels.")
else:
    print("Both groups show similar relative spread in antibody levels.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].hist(antibody_levels_under_40, bins=10, alpha=0.7, label='Under 40')
ax[0].hist(antibody_levels_over_60, bins=10, alpha=0.7, label='Over 60')
ax[0].set_title('Overlapping Histograms of Antibody Levels by Age Group')
ax[0].set_xlabel('Antibody Level')
ax[0].set_ylabel('Frequency')
ax[0].legend()
ax[0].grid(True, axis='y')

# Boxplot
sns.boxplot(data=[antibody_levels_under_40, antibody_levels_over_60], ax=ax[1])
ax[1].set_xticklabels(['Under 40', 'Over 60'])
ax[1].set_title('Boxplot of Antibody Levels by Age Group')
ax[1].set_xlabel('Age Group')
ax[1].set_ylabel('Antibody Level')
ax[1].grid(True, axis='y')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

new_users = np.array([55, 60, 58, 120, 70, 65, 62, 59, 200, 68, 59, 75, 58, 66, 64, 61, 63, 130, 67, 60, 58, 62, 59, 66, 120, 61, 60, 65, 63, 70])
returning_users = np.array([85, 90, 88, 87, 86, 89, 91, 88, 86, 90, 89, 85, 92, 88, 90, 87, 89, 90, 88, 86, 89, 90, 91, 92, 87, 88, 89, 91, 90, 92])

# CV = SD / Mean
def compute_cv(data):
    return np.std(data) / np.mean(data)

cv_new = compute_cv(new_users)
cv_returning = compute_cv(returning_users)

print(f"CV - New Users: {cv_new:.2f}")
print(f"CV - Returning Users: {cv_returning:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram
plt.hist(new_users, bins=10, alpha=0.6, label='New Users')
plt.hist(returning_users, bins=10, alpha=0.6, label='Returning Users')
plt.title('Daily Cart Value Distribution')
plt.xlabel('Cart Value ($)')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Boxplot
sns.boxplot(data=[new_users, returning_users], orient='h')
plt.yticks([0, 1], ['New Users', 'Returning Users'])
plt.title('Boxplot of Cart Values')
plt.xlabel('Cart Value ($)')
plt.show()
```

```python
def five_number_summary(data):
    return {
        "Min": np.min(data),
        "Q1": np.percentile(data, 25),
        "Median": np.median(data),
        "Q3": np.percentile(data, 75),
        "Max": np.max(data)
    }

print("Five-Number Summary - New Users:", five_number_summary(new_users))
print("Five-Number Summary - Returning Users:", five_number_summary(returning_users))
```

```python
import pandas as pd

file_path = "/content/drive/MyDrive/data analysis/daily_sales.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert 'date' to datetime
df['date'] = pd.to_datetime(df['date'])

# Split the data by category
cat_a = df[df['category'] == 'Category A']
cat_b = df[df['category'] == 'Category B']

# Set seaborn style
sns.set(style="whitegrid")

# Create boxplots and histograms for total_sales
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Distribution of Total Sales by Category', fontsize=16)

# Boxplots
sns.boxplot(data=cat_a, y='total_sales', ax=axes[0, 0], color="skyblue")
axes[0, 0].set_title('Boxplot - Category A')
axes[0, 0].set_ylabel('Total Sales (USD)')

sns.boxplot(data=cat_b, y='total_sales', ax=axes[0, 1], color="lightgreen")
axes[0, 1].set_title('Boxplot - Category B')
axes[0, 1].set_ylabel('Total Sales (USD)')

# Histograms
sns.histplot(cat_a['total_sales'], kde=True, bins=15, ax=axes[1, 0], color="skyblue")
axes[1, 0].set_title('Histogram - Category A')
axes[1, 0].set_xlabel('Total Sales (USD)')

sns.histplot(cat_b['total_sales'], kde=True, bins=15, ax=axes[1, 1], color="lightgreen")
axes[1, 1].set_title('Histogram - Category B')
axes[1, 1].set_xlabel('Total Sales (USD)')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

Boxplot Insights:

Category A has lower median sales and a narrower interquartile range (IQR), showing more consistent sales.

Category B displays higher median sales with a larger spread, indicating more variability and a few notable high-value outliers.

Histogram Insights:

Category A's sales distribution is fairly symmetrical and tightly clustered.

Category B shows a right-skewed distribution, suggesting a few days with exceptionally high sales, possibly due to large single transactions or promotions.

```python
# Time series plot of daily total sales for both categories

plt.figure(figsize=(14, 6))
sns.lineplot(data=cat_a, x='date', y='total_sales', label='Category A', marker='o')
sns.lineplot(data=cat_b, x='date', y='total_sales', label='Category B', marker='o')
plt.title('Daily Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales (USD)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```


Category A has relatively stable sales with some periodic dips and peaks.

Category B shows more dramatic changes, with occasional sharp spikeslikely driven by high-value purchases or promotional days.

A few outliers (especially in Category B) could indicate flash sales, campaigns, or abnormal buying behavior.

```python
# Calculate Coefficient of Variation for each category

cv_data = {
    'Metric': ['Total Sales', 'Avg Basket'],
    'Category A CV': [
        cat_a['total_sales'].std() / cat_a['total_sales'].mean(),
        cat_a['avg_basket'].std() / cat_a['avg_basket'].mean()
    ],
    'Category B CV': [
        cat_b['total_sales'].std() / cat_b['total_sales'].mean(),
        cat_b['avg_basket'].std() / cat_b['avg_basket'].mean()
    ]
}

cv_df = pd.DataFrame(cv_data)
cv_df.style.format({
    "Category A CV": "{:.2%}",
    "Category B CV": "{:.2%}"
})
```

```python
# 7 day moving average plot

df_sorted = df.sort_values('date')

# Calculate 7-day moving average for each category
df_sorted['rolling_avg_sales'] = df_sorted.groupby('category')['total_sales'].transform(lambda x: x.rolling(window=7, min_periods=1).mean())

# Separate data by category
cat_a_rolling = df_sorted[df_sorted['category'] == 'Category A']
cat_b_rolling = df_sorted[df_sorted['category'] == 'Category B']

# Plotting the moving averages
plt.figure(figsize=(14, 6))
sns.lineplot(data=cat_a_rolling, x='date', y='rolling_avg_sales', label='Category A (7-Day Moving Avg)')
sns.lineplot(data=cat_b_rolling, x='date', y='rolling_avg_sales', label='Category B (7-Day Moving Avg)')
plt.title('7-Day Moving Average of Daily Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('7-Day Moving Average Total Sales (USD)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='transaction_amount', hue='segment', fill=True, common_norm=False, palette='muted', alpha=0.6)
plt.title('KDE Plot of Transaction Amount by Segment')
plt.tight_layout()
plt.show()
```

```python

summary = df.groupby('segment')['transaction_amount'].agg(['mean', 'median', 'std', 'count'])
summary['IQR'] = df.groupby('segment')['transaction_amount'].quantile(0.75) - df.groupby('segment')['transaction_amount'].quantile(0.25)
summary['CV'] = summary['std'] / summary['mean']
summary = summary[['mean', 'median', 'IQR', 'CV']]
print(summary)
```

```python

plt.figure(figsize=(10, 6))
sns.boxplot(x='segment', y='transaction_amount', data=df, palette='Set2')
sns.swarmplot(x='segment', y='transaction_amount', data=df, color='black', alpha=0.6)
plt.title('Boxplot with Swarm Overlay of Transaction Amount by Segment')
plt.tight_layout()
plt.show()
```
